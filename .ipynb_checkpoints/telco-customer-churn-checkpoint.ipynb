```python
import matplotlib.pyplot as plt

def plot_roc_curve(fpr, tpr):
    """
    Plots a ROC curve
    """
    plt.plot(fpr, tpr, color="orange", label="ROC")
    plt.plot([0,1], [0,1], color="darkblue", linestyle="--", label="Guessing")
    
    plt.xlabel("False positive rate")
    plt.ylabel("True positive rate")
    plt.title("ROC")
    plt.legend()
    plt.show()
    
plot_roc_curve(fpr, tpr)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score

def evaluate_preds(y_true, y_preds):
    accuracy = accuracy_score(y_true, y_preds)
    precision = precision_score(y_true, y_preds)
    recall = recall_score(y_true, y_preds)
    f1 = f1_score(y_true, y_preds)
    metric_dict = {"accuracy": round(accuracy, 2),
                  "precision": round(precision, 2),
                   "recall": round(recall, 2),
                   "f1": round(f1, 2),
                  }
    print(metric_dict)
    return metric_dict
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer


np.random.seed(42)
data = pd.read_csv('./WA_Fn-UseC_-Telco-Customer-Churn.csv')
data = data.drop('customerID', axis=1)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['Churn'] = data['Churn'].map({'Yes': 1, 'No': 0})
data_shufled = data.sample(frac=1)

X = data_shufled.drop('Churn', axis=1)
y = data_shufled['Churn']

categorical_features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']
one_hot = OneHotEncoder(sparse_output=False)
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)], remainder='passthrough')
X_transformed = transformer.fit_transform(X)

# Čuvanje imena kolona
encoded_feature_names = transformer.named_transformers_['one_hot'].get_feature_names_out(categorical_features)
all_feature_names = list(encoded_feature_names) + [col for col in X.columns if col not in categorical_features]

X_transformed_df = pd.DataFrame(X_transformed, columns=all_feature_names)


train_split = round(.7 * len(data_shufled))
valid_split = round(train_split  + .15 * len(data_shufled))

X_train, y_train = X_transformed[:train_split], y[:train_split]
X_valid, y_valid = X_transformed[train_split:valid_split], y[train_split: valid_split]
X_test, y_test = X_transformed[valid_split:], y[valid_split:]

X_train = pd.DataFrame(X_train, columns=all_feature_names)
X_valid = pd.DataFrame(X_valid, columns=all_feature_names)
X_test = pd.DataFrame(X_test, columns=all_feature_names)

X_train['TotalCharges'].fillna(X_train['TotalCharges'].mean(), inplace=True)
X_valid['TotalCharges'].fillna(X_valid['TotalCharges'].mean(), inplace=True)
X_test['TotalCharges'].fillna(X_test['TotalCharges'].mean(), inplace=True)


clf = RandomForestClassifier()
clf.fit(X_train, y_train)
importance_df = pd.DataFrame({'Feature': all_feature_names, 'Importance': clf.feature_importances_})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# Make baseline predictions
y_preds = clf.predict(X_valid)

# Evaluate clasiffer on validatoin set
baseline_metrics = evaluate_preds(y_valid, y_preds)
baseline_metrics, importance_df
```

```python
# Under/over sampling

from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler

np.random.seed(42)
data = pd.read_csv('./WA_Fn-UseC_-Telco-Customer-Churn.csv')
data = data.drop('customerID', axis=1)



data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['Churn'] = data['Churn'].map({'Yes': 1, 'No': 0})
data_shufled = data.sample(frac=1)

indices_no = data_shufled[data_shufled['Churn'] == 0].index
indices_yes = data_shufled[data_shufled['Churn'] == 1].index

# Izdvajanje poduzorka uzoraka za dominantnu klasu ('No')
undersampled_indices_no = np.random.choice(indices_no, len(indices_yes), replace=False)

# Spajanje indeksa poduzorka sa indeksima za klase 'Yes'
undersampled_indices = np.concatenate([undersampled_indices_no, indices_yes])

# Formiranje uravnoteženog DataFrame-a
data_shufled = data_shufled.loc[undersampled_indices]
data_shufled

X = data_shufled[['TotalCharges', 'MonthlyCharges']]
y = data_shufled['Churn']

train_split = round(.7 * len(data_shufled))
valid_split = round(train_split  + .15 * len(data_shufled))

X_train, y_train = X[:train_split], y[:train_split]
X_valid, y_valid = X[train_split:valid_split], y[train_split: valid_split]
X_test, y_test = X[valid_split:], y[valid_split:]

X_train['TotalCharges'].fillna(X_train['TotalCharges'].mean(), inplace=True)
X_valid['TotalCharges'].fillna(X_valid['TotalCharges'].mean(), inplace=True)
X_test['TotalCharges'].fillna(X_test['TotalCharges'].mean(), inplace=True)

print(len(X_train), len(y_train), len(X_test), len(y_test))

clf = RandomForestClassifier()
clf.fit(X_train, y_train)
# importance_df = pd.DataFrame({'Feature': all_feature_names, 'Importance': clf.feature_importances_})
# importance_df = importance_df.sort_values(by='Importance', ascending=False)

# # Make baseline predictions
y_preds = clf.predict(X_test)

# # Evaluate clasiffer on validatoin set
baseline_metrics = evaluate_preds(y_test, y_preds)
clf.feature_importances_
# baseline_metrics, importance_df


from sklearn.metrics import confusion_matrix


cfs = confusion_matrix(y_test, y_preds)

pd.crosstab(y_test, y_preds, rownames=["Actual label"], colnames=["Predicted Labels"])
```

```python
# make predictions

from sklearn.model_selection import GridSearchCV, train_test_split

np.random.seed(42)

data = pd.read_csv('./WA_Fn-UseC_-Telco-Customer-Churn.csv')
data = data.drop('customerID', axis=1)
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['Churn'] = data['Churn'].map({'Yes': 1, 'No': 0})
data_shufled = data.sample(frac=1)

X = data_shufled.drop('Churn', axis=1)
y = data_shufled['Churn']

categorical_features = ['gender', 'SeniorCitizen', 'Partner', 'Dependents', 'tenure',
       'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
       'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
       'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']
one_hot = OneHotEncoder(sparse_output=False)
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)], remainder='passthrough')
X_transformed = transformer.fit_transform(X)

# Čuvanje imena kolona
encoded_feature_names = transformer.named_transformers_['one_hot'].get_feature_names_out(categorical_features)
all_feature_names = list(encoded_feature_names) + [col for col in X.columns if col not in categorical_features]

X = pd.DataFrame(X_transformed, columns=all_feature_names)

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.2)

# X_train = pd.DataFrame(X_train, columns=all_feature_names)
# X_valid = pd.DataFrame(X_valid, columns=all_feature_names)
# X_test = pd.DataFrame(X_test, columns=all_feature_names)

X_train['TotalCharges'].fillna(X_train['TotalCharges'].mean(), inplace=True)
X_test['TotalCharges'].fillna(X_test['TotalCharges'].mean(), inplace=True)
pickle_y_preds = loaded_model.predict(X_test)
```
